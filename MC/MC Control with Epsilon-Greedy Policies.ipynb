```python
def make_epsilon_greedy_policy(Q, epsilon, nA):
    """
    Creates an epsilon-greedy policy based on a given Q-function and epsilon.
    
    Args:
        Q: A dictionary that maps from state -> action-values.
            Each value is a numpy array of length nA (see below)
        epsilon: The probability to select a random action . float between 0 and 1.
        nA: Number of actions in the environment.
    
    Returns:
        A function that takes the observation as an argument and returns
        the probabilities for each action in the form of a numpy array of length nA.
    
    """
    #score, dealer_score, usable_ace = observation 
    def policy_fn(observation):
        a_star = np.argmax(Q[observation])
        ret = []
        for i in range(nA):
            ret.append((1-epsilon + epsilon/nA) if i == a_star else epsilon/nA)
        return ret
    return policy_fn
```

```python
def get_id_first_appearance(state, action, episode):
    for i in range(len(episode)):
        if state == episode[i][0] and action == episode[i][1]:
            return i
```

```python
def mc_control_epsilon_greedy(env, num_episodes, discount_factor=1.0, epsilon=0.1):
    """
    Monte Carlo Control using Epsilon-Greedy policies.
    Finds an optimal epsilon-greedy policy.
    
    Args:
        env: OpenAI gym environment.
        num_episodes: Number of episodes to sample.
        discount_factor: Lambda discount factor.
        epsilon: Chance the sample a random action. Float betwen 0 and 1.
    
    Returns:
        A tuple (Q, policy).
        Q is a dictionary mapping state -> action values.
        policy is a function that takes an observation as an argument and returns
        action probabilities
    """
    
    # Keeps track of sum and count of returns for each state
    # to calculate an average. We could use an array to save all
    # returns (like in the book) but that's memory inefficient.
    returns_sum = defaultdict(float)
    returns_count = defaultdict(float)
    
    # The final action-value function.
    # A nested dictionary that maps state -> (action -> action-value).
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    
    # The policy we're following
    policy = make_epsilon_greedy_policy(Q, epsilon, env.action_space.n)
    
    # Implement this!
    for i in range(num_episodes):
        episode = []
        done = False
        state = env.reset()
        while not done:
            action = np.argmax(policy(state))
            next_state, reward, done, _ = env.step(action)
            episode.append((state, action, reward))
            state = next_state
        
        states_in_episode = set([tuple((x[0], x[1])) for x in episode])
        for state,action in states_in_episode:
            first_id = get_id_first_appearance(state, action, episode)
            G = sum([x[2] for x in episode[first_id:]])
            returns_sum[(state, action)] += G
            returns_count[(state, action)] += 1
            Q[state][action] = returns_sum[(state, action)]/returns_count[(state, action)]
    return Q, policy
```

```python
Q, policy = mc_control_epsilon_greedy(env, num_episodes=500000, epsilon=0.1)
```

```python
# For plotting: Create value function from action-value function
# by picking the best action at each state
V = defaultdict(float)
for state, actions in Q.items():
    action_value = np.max(actions)
    V[state] = action_value
plotting.plot_value_function(V, title="Optimal Value Function")
```
